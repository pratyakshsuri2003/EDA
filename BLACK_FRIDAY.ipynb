now we need to make some preprocessing changes on categorical data

Categorical Features are-
1. Gender
2. Age
3. City_Category
4. Stay_in_current_City_Years

GENDER

```python
dff['Gender'].value_counts()
```

```python
dff['Gender'].unique()
```

```python
dff['Gender'] = dff['Gender'].map({'F': 0, 'M': 1})
dff.head()
```

AGE

```python
dff['Age'] = dff['Age'].map({'0-17': 1, '18-25': 2, '26-35': 3, '36-45': 4, '46-50': 5, '51-55': 6, '55+': 7})
dff.head()
```

City_Category

```python
dff['City_Category'].unique()
```

```python
dff['City_Category'].value_counts()
```

```python
dff_city = pd.get_dummies(dff['City_Category'])
dff_city.head()
```

```python
dff = pd.concat([dff, dff_city], axis=1)
dff.head()
```

```python
dff.drop(['City_Category'], axis=1, inplace=True)
dff.head()
```

```python
# 2 categories of dummies is suffiecient to represent 3 cateogries that's why we delete 'A'
dff.drop(['A'], axis=1, inplace=True)
```

```python
dff.columns
```

Stay_In_Current_City_Years

```python
dff['Stay_In_Current_City_Years'].unique()
```

```python
dff['Stay_In_Current_City_Years'] = dff['Stay_In_Current_City_Years'].str.replace("+", "")
dff.head(10)
```

```python
dff.info()
```

```python
dff['Stay_In_Current_City_Years'] = dff['Stay_In_Current_City_Years'].astype(int)
dff['B'] = dff['B'].astype(int)
dff['C'] = dff['C'].astype(int)
dff.info()
```

```python
dff.drop(['Product_ID'], axis=1, inplace=True)
dff.info()
```
